## Question 1

Import the dataset personality_scores.csv. Examine the data frame for duplicates (based on ID), and drop any duplicates that exist. Use an assert statement to check that the new data frame is the length of the unique entries of the original data frame.

```python
# load module
import pandas as pd
import re
```

```python
# import data set
data = pd.read_csv("personality_scores.csv", sep=';')
```

```python
# checking for missing values
data.isnull().sum().sum()
```

There are 29 545 missing values.

```python
# creating data frame
#df = pd.DataFrame(data)
```

```python
# inspecting duplicate values
data.nunique()
```

```python
# drop duplicate rows and null values
 
data_new = data.drop_duplicates(subset="ID", keep="first").dropna(axis=1)
data_new
```

```python
# assertion
try:
    assert data['ID'].unique().sum() == data_new['ID'].sum(), "the new data frame is not equal to the length of the unique entries of the original data frame"
    print('True')
except AssertionError as msg:
    print(msg)
```

## Question 2

Create new columns containing the total score of each of the personality test subscales. To do this, write a function (or functions) that will calculate the total score for each of the subscales (conscientiousness, emotional stability, openness to new experience, agreeableness, extraversion), as set out in scoring. In other words, for the conscientiousness total score, all items marked as belonging to ‘conscientiousness’ should be summed.

```python
# set ID column as index
#data_new = data_new.set_index('ID')

# Making dictionary
category= dict({1:'Extraversion', 2:'Agreeableness', 
              3:'Conscientiousness', 4:'Emotional Stability', 
              5:'Intellect'})

# set the function that calculates the total score for each subscale
def score(entry, category):    
    '''
    The function calculates the total 
    score for each of the subscales.
    '''
    numbers=0
    r = re.compile('\d+')
    for col in entry:
        num= r.findall(col)
        if str(category) == num[0]:
            numbers+= int(num[1])
        else:
            numbers= numbers
    return numbers
    
for i in category.keys():
    data_new[f'{category[i]}']= data_new.loc[:,'Section 5 of 6 [I am always prepared.]':
                                   'Section 5 of 6 [I worry about things.]'].apply(score,args=(i,), axis=1)
```
