## QUESTION 2

### Total score of each of the personality test subscales

```python
# Making dictionary
category = dict({1:'Extraversion', 2:'Agreeableness', 3:'Conscientiousness', 4:'Emotional Stability', 5:'Intellect'})

# set the function that calculates the total score for each subscale
def score(entry, category):
    '''
    The function calculates the total 
    score for each of the subscales.
    '''
    numbers=0
    _r= re.compile('\d+')
    for col in entry:
        num= _r.findall(col)
        if str(category)==num[0]:
            numbers+= int(num[1])
        else:
            numbers= numbers
    return numbers
    
for i in category.keys():
    df_score[f'{category[i]}']= df_score.loc[:,'Section 5 of 6 [I am always prepared.]':'Section 5 of 6 [I worry about things.]'].apply(score, args=(i,) , axis=1)
    
df_score.head()
```

## QUESTION 3

### Load and read in department Data Frame

```python
df_dep = pd.read_csv("departments.csv", sep=";")
df_dep.head()
```

### Merging the personality score and the department data frame

```python
#merging department with personality scores
df_merged = pd.merge(df_score, df_dep ,how='left')
df_merged.head()
```

###  Assert statement to check that the newly created merged data frame has the same amount of rows as the department data frame, and the expected number of columns 

```python
try:
    assert len(df_score) == len(df_merged), "The two data frames do not have the same amount of rows" 
    print('True')
except AssertionError as msg:
    print(msg)
```

```python
# Assert that the merged data frame has the expected number of columns
try:
    assert len(df_merged.columns) != len(df_score.columns), "The merged data frame does not return the expected number of columns"
    print('True')
except AssertionError as msg:
    print(msg)
```

### Filter the merged data frame so that you get only the applicants who scored less than 30 on emotional stability, conscientiousness AND agreeableness.

```python
df_merged['Risk']=np.where((df_merged['Conscientiousness']<30) & 
                          (df_merged['Agreeableness']<30) & 
                          (df_merged['Emotional Stability'] <30),'high risk','low risk')

df_merged.head()
```
